```python
#Data Analysis
#1. What products have the highest profit margin?

df1=df.groupby('Category')['Profit'].sum()
df1.plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

```python
#Data Analysis
#2. What is the region with the highest sales?

df2=df.groupby('Region').sum()['Sales']
df2.plot(kind='bar')
plt.title('Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Data Analysis 
#3. Which month have the highest sales?

df['Order Month']=pd.DatetimeIndex(df['Order Date']).month
df3=df.groupby('Order Month').sum()['Sales']
df3.plot(kind='line')
plt.title('Total Sales per Monthly Basis')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Data Analysis 
#4. Which Shipping mode have the highest amount of return rate?

df5=df.groupby('Ship Mode').size()
df5_1=df[df['Profit']<0].groupby('Ship Mode').size()
df5_2=(df5/df5_1)*100
df5_2.plot(kind="bar")
plt.title("Return By Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Return Rate")
plt.show()
```

```python
#Data Analysis 
#5. What are the days of the week that have the lowest profits?

df['Order day']=pd.DatetimeIndex(df['Order Date']).day_name()
df5=df.groupby('Order day').sum()['Profit']
df5.plot(kind='bar')
plt.title('Return Rate per Shipping Mode')
plt.xlabel('Day of the Week')
plt.ylabel('Total Profit')
plt.show()
```
